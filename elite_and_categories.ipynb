```python
#expect 30 seconds for this shit to finish
import pandas as pd
df = pd.read_csv("yelpdata/yelp-csv/yelp_academic_dataset_business.csv")

df_categories = df[['business_id', 'categories']]

business_id_labels = list() 
business_id_done = set() 


for i in df_categories.index: 
    business_id = df_categories.loc[i, 'business_id']
    categ_string = df_categories.loc[i, 'categories']
    if(str(categ_string) != "nan"): 
        categ_list = categ_string.split(', ')
        if not ( business_id in business_id_done) : 
            business_id_done.add(business_id)
            business_id_labels.extend(list(map(lambda label : (business_id, label) ,categ_list)))
            

business_id_labels_df = pd.DataFrame(business_id_labels, columns = ['business_id', 'businessLabel'])
business_id_labels_df.to_csv("business_id_categories.csv", index = False)
print('done')
```
