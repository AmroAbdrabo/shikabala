```python
import pandas as pd
df = pd.read_csv('/Users/amroa/Downloads/yelpdata/yelp-csv/yelp_academic_dataset_business.csv')
df.dropna(subset = ['business_id','stars', 'review_count', 'name', 'is_open'], inplace= True)
df
```

```python
import numpy as np
df = pd.DataFrame(df[['business_id','attributes']])
df = df.dropna()
df
```

```python
filtamb = df['attributes'].str.contains("Noise")
df = df[filtamb]
df['attributes']
```

```python
df.loc[18, 'attributes']
```

```python
weights = {'very_loud':3, 'loud': 2, 'average': 1, 'quiet': 0}
def attributeNoiseToAid(attributes):
    splitAtAmbience = str(attributes).split("'NoiseLevel': ")
    if((splitAtAmbience[1]).startswith("'None'")):
        return -1
    beginIndex      = splitAtAmbience[1].find("'")
    endIndex        = splitAtAmbience[1].find("'", beginIndex + 1)
    mynoise = splitAtAmbience[1]
    return weights[mynoise[(beginIndex+1):endIndex]]
```

```python
df['attributes'] = df['attributes'].apply(attributeNoiseToAid)
```

```python
df.rename(columns = {'attributes': 'noiseLevel'}, inplace = True)
```
