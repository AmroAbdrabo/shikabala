```python
import pandas as pd
df = pd.read_csv('/Users/amroa/Downloads/yelpdata/yelp-csv/yelp_academic_dataset_business.csv')
df.dropna(subset = ['business_id','stars', 'review_count', 'name', 'is_open'], inplace= True)
```

```python
import numpy as np
df = pd.DataFrame(df[['business_id','attributes']])
df = df.dropna()
df
```

```python
filtamb = df['attributes'].str.contains("BusinessParking")
df = df[filtamb]
df['attributes']
```

```python
weights = {'garage': 81,
          'street': 27, 'validated': 9, 'lot': 3, 'valet': 1}
def attributeMusicToAid(attributes):
    splitAtAmbience = str(attributes).split("'BusinessParking': ")
    beginIndex      = splitAtAmbience[1].find('{')
    endIndex        = splitAtAmbience[1].find('}')
    ambience_fields = (splitAtAmbience[1])[beginIndex:(endIndex+1)]
    splitted_fields = ambience_fields.split(': ')
    
    #trim out the annoying braces
    splitted_fields[0] = (splitted_fields[0])[1:]
    splitted_fields[-1] = (splitted_fields[-1])[0:4]
    if((splitAtAmbience[1]).startswith("'None'")):
        return -1
    
    # get the aid
    aid = 0
    for i in range(0, len(splitted_fields) -1):
        commasplitted1 = splitted_fields[i].split(', ')
        commasplitted2 = splitted_fields[i+1].split(', ')
        name_fields = commasplitted1[-1] #e.g. touristy
        name_value = 2*int(commasplitted2[0] == 'True') + 1*int(commasplitted2[0] == 'False')
        aid = aid + weights[name_fields[1: -1]]*name_value
    return aid
```

```python
df['attributes'] = df['attributes'].apply(attributeMusicToAid)
```

```python
df.replace(-1, np.nan, inplace=True)
df = df.dropna()
df.rename(columns = {'attributes': 'PID'}, inplace = True)
```

```python
df['PID'] = df['PID'].astype('int')
df
```

```python
df.to_csv("offersParkingAB.csv", index = False)
df_parking = pd.DataFrame(df['PID'])
df_parking
```

```python
#map the ternary strings to list
def bitcountToRomBoolean(aid, ambfield):
    leftShift = aid//weights[ambfield]
    value__ = (leftShift)%3
    return value__
    
df_parking['garage'] = df_parking['PID'].apply(lambda x : bitcountToRomBoolean(x, 'garage'))
df_parking['street'] = df_parking['PID'].apply(lambda x : bitcountToRomBoolean(x, 'street'))
df_parking['validated'] = df_parking['PID'].apply(lambda x : bitcountToRomBoolean(x, 'validated'))

df_parking['lot'] = df_parking['PID'].apply(lambda x : bitcountToRomBoolean(x, 'lot'))
df_parking['valet'] = df_parking['PID'].apply(lambda x : bitcountToRomBoolean(x, 'valet'))
```

```python
#remove duplicates
df_parking.drop_duplicates(subset ="PID", 
                     keep = 'first', inplace = True) 
df_parking
```

```python
df_parking.to_csv("parkingAB.csv", index = False)
```
