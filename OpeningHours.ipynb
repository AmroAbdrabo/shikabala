```python
import pandas as pd
df = pd.read_csv('yelpdata/yelp-csv/yelp_academic_dataset_business.csv')
df
```

```python
df = pd.DataFrame(df[['business_id','hours']], columns = ['business_id', 'hours'])
df = df.dropna()
df
```

```python
#as always parse a sample
df.loc[1]['hours']
```

```python
day_idx = {'Monday':0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 'Friday': 4, 'Saturday':5, 'Sunday':6}
df_hasOpeningTimes = pd.DataFrame(columns = ['business_id', 'dayIndex', 'timeTo', 'timeFrom'])
df_hasOpeningTimes
```

```python
new_rows = list()
def parseJSON(business, hours):
    for day in day_idx:
        idx = hours.find(day)
        if (idx > 0):
            idx_start = hours.find("'", idx+len(day)+1)
            idx_end = hours.find("'", idx_start+1)
            hoursparsed = hours[(idx_start+1):idx_end]
            splitathyphen = hoursparsed.split('-') 
            timeFrom = splitathyphen[0]
            timeTo = splitathyphen[1]
            minutehours1 = timeFrom.split(':')
            minutehours2 = timeTo.split(':')
            hours_in_minutes1 = int(minutehours1[0])*60+int(minutehours1[1])
            hours_in_minutes2 = int(minutehours2[0])*60+int(minutehours2[1])
            new_rows.append([business, day_idx[day], hours_in_minutes2, hours_in_minutes1])
            
        
#check for garbage (elhamdulillah, there is none)
#for index, row in df.iterrows():
#    if ('NAN' in str(row['hours']).upper()):
#        print(row)
#print('done')
    
```

```python
for index, row in df.iterrows():
    parseJSON(row['business_id'], row['hours'])
```

```python
new_rows
```

```python
df_hasOpeningTimes = pd.DataFrame(new_rows, columns = ['business_id', 'dayIndex', 'timeTo', 'timeFrom'])
df_hasOpeningTimes
#save df_hasOpeningTimes as csv
```
