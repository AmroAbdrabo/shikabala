```python
import pandas as pd
df = pd.read_csv('yelpdata/yelp-csv/yelp_academic_dataset_business.csv')
df
```

```python
#search for garbage nan in longitude and latitude (#garbage nans exist)
for index, row in df.iterrows():
    if ('NAN' in str(row['longitude']).upper()):
        print('garbage present')
    if ('NAN' in str(row['latitude']).upper()):
        print('garbage present')
    if ('NAN' in str(row['city']).upper()):
        print(str(row) + " with index " + str(index))
        print('garbage present in city')
    if ('NAN' in str(row['state']).upper()):
        print('garbage present in state')
print('done')
```
